```python
from urllib import request
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
# request.urlretrieve('https://i.imgur.com/dDOnan4.jpg', './test.jpg')
```

```python
import time
import random

for t in tmp_list:
    title = t['title']
    article_url = 'https://www.dcard.tw/f/photography/p/'+ str(t['id'])
    print(title)
    print(article_url)
    for n, url_dict in enumerate(t['mediaMeta']):
        img_url = url_dict['url']
        print(img_url)
        try:
            request.urlretrieve(img_url, './dcard_img/%s_%s.jpg'%(title, n))
        except HTTPError as e:
            print(n, img_url)
            print(e.args)
        except FileNotFoundError as e:
            print(n, img_url)
            print(e.args)
            tmpname = img_url.split('/')[-1]
            request.urlretrieve(img_url, './dcard_img/%s'%(tmpname))
        s = random.randint(1,5)
        time.sleep(s)
    print()
```
