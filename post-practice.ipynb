# 正式操作

```python
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}

url = 'https://buzzorange.com/techorange/wp-admin/admin-ajax.php'


for page in range(0, 10):
    data = {'action': 'fm_ajax_load_more',
           'nonce': '543a6a8482',
           'page': page+1}

    res = requests.post(url, headers=headers, data=data)
    html = json.loads(res.text)['data']
    soup = BeautifulSoup(html, 'html.parser')
    for t in soup.select('h4 a'):
        print(t.text)
        print(t['href'])
        print()
```
