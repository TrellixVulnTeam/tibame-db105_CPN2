```python
import jieba
import os
```

```python
filename_list = os.listdir('./res_gossiping')
filename_list
```

```python
article_str = ''
for article in filename_list:
    with open('./res_gossiping/%s'%(article), 'r', encoding='utf-8') as f:
        tmpstr = f.read().split('---split---')[0]
        article_str += tmpstr
#     print(tmpstr)
#     print('===========================')
article_str
```

```python
cut_list = jieba.cut(article_str)
word_list = list(cut_list)
```

```python
word_count = {}
for w in word_list:
    if len(w) > 1:
        if w in word_count:
            word_count[w] += 1
        else:
            word_count[w] = 1
word_count
```

```python
[(k, word_count[k]) for k in word_count]
```

```python
list123 = []
for k in word_count:
    tmp_tuple = (k, word_count[k])
    list123.append(tmp_tuple)
list123
```

```python
list123.sort(key=lambda x: x[1], reverse=True)
```
