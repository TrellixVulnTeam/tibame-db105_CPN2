```python
data_list = []
for article in filename_list:
    with open('./res_gossiping/%s'%(article), 'r', encoding='utf-8') as f:
        tmpstr = f.read().split('---split---')[1]
        tmplist = tmpstr.split('\n')[1:-1]
        data_list.append(tmplist)
#     print(tmplist)
#     print('===========================')
data_list
```

```python
df2 = pd.DataFrame(data_list, columns=['推', '噓', '分數', '作者', '標題', '時間'])
df2
```

```python
def data_filter(tmpstr):
    output = tmpstr.split(': ')[1]
    return output
```

```python
df2['推']
```

```python
df2['推'].apply(data_filter)
```

```python
df2['推'] = df2['推'].apply(data_filter)
df2
```

```python
df_bak = df2
```

```python
for i in df2:
    try:
        df2[i] = df2[i].apply(data_filter)
    except IndexError:
        pass
```
